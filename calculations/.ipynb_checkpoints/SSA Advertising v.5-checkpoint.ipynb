### Огибающая кривая
Для того, чтобы понять характер сезонности (аддитивность, мультипликативность) построим огибающую кривую для исходного ряда и для логарифма исходного ряда. <br>
Для этого, прежде всего, извлечем трендовую составлящую из ряда с помощью SSA.

```r
impressions_ts <- ts(impressions[25:3384,4], frequency = 7*24)
```

```r
s <- ssa(impressions_ts, L = 168)
r_1 <- reconstruct(s, groups = list(trend = c(1)))
plot(r_1)
```

Огибающая кривая для исходного ряда (после вычета тренда)

```r
r_new <- impressions_ts - r_1$trend
# plot(impressions_ts[,1] - r_0$trend)
r_new_2 <- 2*(r_new^2)
s <- ssa(r_new_2, L = 168)
r_1_2 <- reconstruct(s, groups = list(trend = c(1)))

plot(r_new, col = 'green')
lines((r_1_2$trend)^(1/2), col = 'red')
```

Теперь возьмем логаримф исходного ряда и повторим процедуру

```r
impressions_ts <- ts(impressions[25:3384,5], frequency = 7*24)
s <- ssa(impressions_ts, L = 168)
r_2 <- reconstruct(s, groups = list(trend = c(1)))
plot(r_2)
```

Огибающая кривая выглядит гораздо более ровной, чем для ряда до логарифмирования (если не учитывать конец ряда). Отсюда делаем вывод о мультипликативном характере сезонности

```r
r_new <- impressions_ts - r_2$trend
# plot(impressions_ts[,1] - r_0$trend)
r_new_2 <- 2*(r_new^2)
s <- ssa(r_new_2, L = 168)
r_2_2 <- reconstruct(s, groups = list(trend = c(1)))

plot(r_new, col = 'green')
lines((r_2_2$trend)^(1/2), col = 'red')
```

### Прогнозирование

Далее попробуем применить различные методы прогнозирования и сравнить результаты <br>
Сравнивать будем по показателю RMSE <br>
Прогноз будем осуществлять скользящим методом <br>
Прогнозировать будем одни сутки (24 значения)

Baseline подходом будет просто значения предыдущего дня

```r
baseline_approach <- c()



for (i in 1:135) {
    start_index <- 24*i+1
    end_index <- 24*(1+i)
    baseline_approach <- c(baseline_approach,
                       accuracy(exp(ts(impressions[(end_index+1):(end_index+24),5], frequency = 7*24)), exp(ts(impressions[start_index:end_index,5], frequency = 7*24)))[,'RMSE']
                       ) 
}
```

Посчитаем целевую метрику: средний RMSE по всем прогнозам

```r
mean(baseline_approach)
```

Посмотрим на график RMSE ошибки в динамике. Видно, что к концу периода ошибка возрастает

```r
plot(baseline_approach, type = 'l')
```

Первый подход. <br>
SSA со стандартной длиной окна равной половине ряда и топ-10 сингулярными тройками <br>
Модель обучается на всех доступных данных

```r
first_approach <- c()

for (i in 1:135) {
    impressions_ts_s <- ts(impressions[25:(24*(3+i)+24),5], frequency = 7*24)
    s <- ssa(impressions_ts_s)
    ssa.for <- rforecast(s, groups = list(trend = c(1:10)),len = 24)
    first_approach <- c(first_approach,
                       accuracy(exp(ts(impressions[(24*(3+i)+24+1):(24*(3+i)+24+24),5], frequency = 7*24)), exp(ts(ssa.for, frequency = 7*24)))[,'RMSE']
                       ) 
}


# cbind(1,accuracy(exp(ts(impressions[3361:3384,5], frequency = 7*24)), exp(ts(ssa.for, frequency = 7*24)))[,'RMSE'])
```

Прогноз по нашей метрике получился хуже, чем базовый

```r
mean(first_approach)
```

```r
plot(first_approach, type = 'l')
```

Теперь попробуем подобрать параметры SSA считая целевую метрику по решетке параметров

```r
moving_forecast <- function(period, window_l, eigens) {
    result <- c()

    if (period == -999) {
        end_of_cycle <- (length(impressions_2[,5]) - 24)/24
    }
    else {
        end_of_cycle <- (length(impressions_2[,5]) - 24 - period)/24
    }
    
    for (i in 1:end_of_cycle) {
        if (period == -999) {
            start_index <- 1
            end_index <- 24*i
        }
        else {
            start_index <- 24*i - 23
            end_index <- period*i
        }
        
        impressions_ts_s <- ts(impressions_2[start_index:end_index,5], frequency = 7*24)
        if (window_l == -999) {
            L <- (length(impressions_ts_s) + 1) %/% 2
        }
        else {L <- window_l}
#         print(paste(i,period, start_index, end_index, L, eigens))
        try(
        {s <- ssa(impressions_ts_s, L = L)
        ssa.for <- rforecast(s, groups = list(trend = c(1:eigens)),len = 24)
        result <- c(result,
           accuracy(exp(ts(impressions_2[(end_index + 1):(end_index + 24),5], frequency = 7*24)), 
                    exp(ts(ssa.for, frequency = 7*24)))[,'RMSE']
           )
        }

        )
        

    }
#     print(result)
#     print('-------')
    return(mean(result, na.rm=TRUE))
}
```

```r
result_df <- data.frame()

for (period in c(168)) {
    for (window_l in c(-999, 24)) {
        for (eigens in c(5)) {
            result_df <- rbind(result_df,data.frame('period' = period, 'window_l' = window_l, 'eigens' = eigens, 'Average RMSE' = moving_forecast(period, window_l, eigens )))
        }
    }
}
```
